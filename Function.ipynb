```python
students = [
    {"name": "Nikhil", "score": 85},
    {"name": "Rahul", "score": 45},
    {"name": "Charan", "score": 75},
    {"name": "Subbu", "score": 60}
]
def res(students):
  passed = filter(lambda x: x["score"] >= 60, students)
  graded = map(lambda x: {**x, "grade": "Pass"}, passed)
  result = sorted(graded, key=lambda x: x["score"], reverse=True)
  return result
print(res(students))
```

```python
from functools import reduce
def calculator(*args,operator):
  if operator == "add":
    return reduce(lambda x,y: x+y,args)
  elif operator == "multi":
    return reduce(lambda x,y: x*y,args)
  elif operator == "sub":
    return reduce(lambda x,y: x-y,args)
  elif operator == "div":
    return reduce(lambda x,y: x % y,args)
  elif operator == "min":
    return min(args)
  elif operator == "max":
    return max(args)
op = input()
print(calculator(10,20,30,operator = op))
```

```python
#nearest prime number before and after
def prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

n = int(input())

before = n - 1
after = n + 1

while not prime(before):
    before -= 1

while not prime(after):
    after += 1

print("Before Prime =", before)
print("After Prime =", after)
```

```python
#nearest prime number
def prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

n = int(input())

left = n - 1
right = n + 1

while not prime(left):
    left -= 1

while not prime(right):
    right += 1

if n - left <= right - n:
    print(left)
else:
    print(right)
```
